# async, sync

```python
import asyncio
import time

async def invoke_async(llm, messages):
    response = await llm.ainvoke(messages)
    print(response.content)

async def invoke_parallel(llm, messages):
    tasks = [invoke_async(llm, messages) for _ in range(10)]
    await asyncio.gather(*tasks)
```

## async

```python
messages = [
    ('system', '너는 서울의 음식과 문화 전문가야.'),
    ('human', '서울을 대표하는 음식을 맛볼 수 있는 레스토랑 5개를 추천해줘.')
]
start = time.perf_counter()

await invoke_parallel(llm, messages)

end = time.perf_counter()

print(f'Elapsed time: {end - start:.2f} seconds')
```

## sync

```python
# sync 호출 시간 비교 5 times
```

```python
start = time.perf_counter()
for _ in range(5):
    response = llm.invoke(messages)
    print(response.content)
end = time.perf_counter()
print(f'Elapsed time: {end - start:.2f} seconds')
```
